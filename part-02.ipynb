# Flu Forecasting with the Help of Social Media Sentiment 







# Specific Goal

The objective is to build an algorithm to combine flu surveillance, data from Google Flu Trends and Twitter to help predict where, peak and severity of influenza outbreak in a given season. 

# Problem Statement

Traditional surveillance methods lag 1-2 weeks and represents a limited syndromic observation of people who seek medical attention each week.

# Risks & Assumptions

    1. Keywords and search engine query capture all relevant episodes
    2. Data processing and preparation and sentiment
    3. Signal-to-noise issue 
    4. Over-prediciton


# Proposed Methods & Models


NLP - signal vs noise with tweets

ARIMAX - comine tweets, google search query, CDC Influenza Like Illness data and CDC Laboratory-confirmed influenza cases
